## CSV Loader

```python
from langchain.document_loaders import CSVLoader
```

```python
loader = CSVLoader(file_path='../datasets/sns_datasets/titanic.csv') # Lazy Loader
loader
```

```python
data = loader.load()

data # Python list of document objects; Each row in a separate document object
```

```python
data[0] # Single Row
```

```python
type(data[0])
```

```python
# To get the document content

print(data[0].page_content)
```

```python
# Metadata

print(data[0].metadata)
```

```python
# Specify a column name to identify the dataset

# data = CSVLoader(file_path='../datasets/sns_datasets/titanic.csv', source_column= 'sex').load()
data = CSVLoader(file_path='../datasets/sns_datasets/titanic.csv', source_column= 'gender').load()

data
```

## Directory Loader

```python
import os
from langchain.document_loaders import DirectoryLoader
```

```python
dirname = "../datasets/harry_potter_mix"
```

```python
loader = DirectoryLoader(path=dirname)
```

```python
data = loader.load()

len(data)
```

```python
len(os.listdir(dirname))
```

```python
# Specify the file formats

loader = DirectoryLoader(path=dirname, glob='**/*.html')

data = loader.load()

len(data)
```
